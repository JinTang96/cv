## Augmented Reality



Augmented reality is a way of improving or ehnacing real world view with superimposed computer generated elements such as images, videos or 3D models.

There are two types of augmented reality **Location-based** and **Recognition-based** augmented reality.

### Location-based   

Relies on detecting the user's location and orientation by reading data from several sensors. This information is used to superimpose computer-generated elements on the screen

### Recognition-based

Uses image processing techniques to identify the place the user is looking.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('../samples/data/opencv_logo_with_text.png')
img = cv.imread('assets/data/opencv_logo_with_text.png')

plt.figure(figsize=(10,6))
# BGR to RGB
imgRGB = img[:,:,::-1]
plt.subplot(1,2,1)
plt.title('original')
plt.imshow(imgRGB)

orb = cv.ORB_create()
keypoints = orb.detect(img, None)

keypoints, descriptors = orb.compute(img, keypoints)

imgKeypoints = cv.drawKeypoints(img, keypoints, None, color=(255,0,255), flags=0)

# BGR to RGB
imgRGB = imgKeypoints[:,:,::-1]
plt.subplot(1,2,2)
plt.title('keypoints')
plt.imshow(imgRGB)

plt.show()
```
