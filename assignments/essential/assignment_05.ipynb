## Class activity
### Question 01 - Object Oriented Programming
1. Define a class of student
2. The class contains there properties: name, matric card number and courses
3. The class also contains some methods:
- __init__ method takes the initial properties from the user to be filled. 
- Print method to print all the information of the student in a good format<br>
4. After a class is defined, construct 3 objects from the class with attributes of the objects.<br>
Use the print function of the class to print the object and its attributes.


```python
class Student:
    # default constructor 
    def __init__(self, name, ID, course):
        self.name = name
        self.ID = ID
        self.course = course

    # show method
    def show(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.ID}")
        print(f"Course: {self.course}\n")

student1 = Student("Alice", "CS10001", ['Physics', 'Biology', 'Chemistry'])
student2 = Student("Joshua", "CS10002", ['Additional Mathematics', 'Science', 'Mandarin'])
student3 = Student("Kelvin", "CS10003", ['English', 'Account', 'Economy'])


student1.show()
student2.show()
student3.show()

# print(student1.__dict__)
# Output: {'name': 'Alice', 'ID': 'CS10001', 'course': ['Physics', 'Biology', 'Chemistry']}
```
