# Assignment

1. Create an array of shape (100, 100, 3) of unsigned integer 8 bits filled with random values between 0 and 256.
2. Find the values that are less than 30, 50, 100, and 250.
3. Split the array into 4 parts vertically and horizontally.
4. Copy array from question 1 and change the values that are less than 110 to 0 and others to 255.
5. Hint : use random seed of 12345

```python
# Question 01
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(67)
a = np.random.randint(0,256, (100,100, 3), dtype='uint8')

print('Shape:', a.shape)
print('Size:', a.size)
print('Dimension:', a.ndim)
print('Type:', a.dtype)

plt.imshow(a)
plt.show()
```

```python
# Question 02
a_30 = a[a < 30]
print('Values less than 30:\n', a_30)

a_50 = a[a < 50]
print('Values less than 50:\n', a_50)

a_100 = a[a < 100]
print('Values less than 100:\n', a_100)

a_250 = a[a < 250]
print('Values less than 250:\n', a_250)
```

```python
# Question 03
a_vertical = np.vsplit(a,4)
print('Array element in vertically splitted array:',len(a_vertical))

a_horizontal = np.hsplit(a,4)
print('Array element in horizontally splitted array:',len(a_horizontal))
```

```python
# Question 04
b = a.copy()

# b[b < 110]= 0
# b[b >= 110]= 255

b_new = np.where(b < 110, 0, 255)
print(b_new)

plt.imshow(b_new)
plt.show()
```
