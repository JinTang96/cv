2. Draw a hexagon
3. Draw a decagon
4. Draw a line with 7 points that will not meet

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# Define color dictionary
colors = {
    'LIGHT_GRAY': (200, 200, 200),
    'BLACK' : (0,0,0),
    'WHITE' : (255,255,255)
    }

# Define a function for creating a canvas 
def create_canvas(height=500, width=500, bgColor=0):
    # create the image (canvas)
    canvas = np.zeros((height, width, 3), dtype='uint8')
    if bgColor != 0:
        canvas[:] = bgColor
    return canvas

# Define a function for displaying
def show_in_matplotlib(canvas, figsize=(6,8)):
    canvasRGB = cv.cvtColor(canvas, cv.COLOR_BGR2RGB)
    plt.figure(figsize=figsize)
    #plt.axis('off')
    plt.imshow(canvasRGB)
    plt.show()
```

```python
# Create canvas
canvas = create_canvas(height=400, width=400, bgColor=colors['WHITE'])

# Define the coordinates of the shapes
triangle = np.array([[100,25],[35,160],[165,160]], np.int32)


hexagon = np.array([[338,35],[263,35],[225,100],
                    [263,165],[337,165],[375,100]], np.int32)  

decagon = np.array([[123,229],[77,229],[39,256],[25,300],[39,344],
                    [77,371],[123,371],[161,344],[175,300],[161,256]], np.int32)
                            
seven_point = np.array([[300,225],[241,253],[227,317],[267,368],
                        [333,368],[373,317],[359,253]], np.int32)

# Draw line divider
cv.line(canvas, (200,0), (200,400), colors['LIGHT_GRAY'], 1)
cv.line(canvas, (0,200), (400,200), colors['LIGHT_GRAY'], 1)

# Draw the shapes
cv.polylines(canvas, [triangle], True, colors['BLACK'], 3)
cv.polylines(canvas, [hexagon], True, colors['BLACK'], 3)
cv.polylines(canvas, [decagon], True, colors['BLACK'], 3)
cv.polylines(canvas, [seven_point], False, colors['BLACK'], 3)

# Point for seven_point
cv.circle(canvas, (300,225), 7, colors['BLACK'], -1)
cv.circle(canvas, (241,253), 7, colors['BLACK'], -1)
cv.circle(canvas, (227,317), 7, colors['BLACK'], -1)
cv.circle(canvas, (267,368), 7, colors['BLACK'], -1)
cv.circle(canvas, (333,368), 7, colors['BLACK'], -1)
cv.circle(canvas, (373,317), 7, colors['BLACK'], -1)
cv.circle(canvas, (359,253), 7, colors['BLACK'], -1)

# Text for each shape
cv.putText(canvas, 'TRIANGLE', (63,125), cv.FONT_HERSHEY_PLAIN, 1, colors['BLACK'])
cv.putText(canvas, 'HEXAGON', (265,110), cv.FONT_HERSHEY_PLAIN, 1, colors['BLACK'])
cv.putText(canvas, 'DECAGON', (63,305), cv.FONT_HERSHEY_PLAIN, 1, colors['BLACK'])
cv.putText(canvas, '7 POINTS', (265,305), cv.FONT_HERSHEY_PLAIN, 1, colors['BLACK'])

show_in_matplotlib(canvas)
```
