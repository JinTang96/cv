```python
# Quesiton 04

import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt

# Load image
path = '../../essential/assets/ca2.jpg'
img = cv.imread(path)

# Change image from BGR to RGB
img = img[:,:,::-1]

# Define types of kernels
laplace_kernel = np.array([[-1, -1, -1],
                              [-1, 8, -1],
                              [-1, -1, -1]])

sobel_x_kernel = np.array([[1, 0, -1],
                           [2, 0, -2],
                           [1, 0, -1]])

sobel_y_kernel = np.array([[1, 2, 1],
                           [0, 0, 0],
                           [-1, -2, -1]])

prewitt_x_kernel = np.array([[1, 0, -1],
                             [1, 0, -1],
                             [1, 0, -1]])

prewitt_y_kernel = np.array([[1, 1, 1],
                             [0, 0, 0],
                             [-1, -1, -1]])

# Apply kernel filter on the image
img_laplace_kernel = cv.filter2D(img, -1, laplace_kernel)
img_sobel_x_kernel = cv.filter2D(img, -1, sobel_x_kernel)
img_sobel_y_kernel = cv.filter2D(img, -1, sobel_y_kernel)
img_prewitt_x_kernel = cv.filter2D(img, -1, prewitt_x_kernel)
img_prewitt_y_kernel = cv.filter2D(img, -1, prewitt_y_kernel)

# Convert filtered images into a list
img_list = [img, img_laplace_kernel, 
            img_sobel_x_kernel, img_sobel_y_kernel, 
            img_prewitt_x_kernel, img_prewitt_y_kernel]
title_list = ['Original image','Laplace kernel',
                'Sobel X kernel','Sobel Y kernel',
                'Prewitt X kernel','Prewitt Y kernel']

# Declare plot parameters
rows = 3
cols = 2
fsize = 20
plt.figure(figsize=(15,15))

# Display all the images
for i in range(len(img_list)):
    
    plt.subplot(rows,cols,i+1)
    plt.imshow(img_list[i])
    plt.title(title_list[i])
    plt.axis('off')
```
